```python
import pandas as pd
```

```python
df = pd.read_csv('../data/partition_predictions/window_topic_prediction_with_short_sentences_merged_54_2016.csv')
```

```python
programs = ['BBC News at Six', 'BBC News at Ten', 'BBC News at One']
df = df.drop(['Unnamed: 0'], axis=1)
df = df.loc[df.source.isin(programs)]
df['date'] = pd.to_datetime(df['date'])
mask = (df['date'] > '2016-1-1') & (df['date'] <= '2016-6-30')
df = df.loc[mask]
```

```python
df.source.unique()
```

```python
df
```

```python
rows = []

last_topic = None
last_partition = None
for index, row in df.iterrows():
    partition_id = row['partition_id']
    dt = row['date']
    source = row['source']
    transcript = row['transcript']
    t = row['type']
    topic = row['topic']
    
    
    if last_partition == partition_id and last_topic == topic:
        rows[-1][3] += row['transcript']
    else:
        last_topic = topic
        last_partition = partition_id
        rows.append([partition_id, dt, source, transcript, t, topic])
res_df = pd.DataFrame(rows, columns=df.columns)
res_df
```

```python
z = res_df.loc[res_df.source == 'BBC News at Ten']
z['date'] = pd.to_datetime(z['date'])
z['month'] = z['date'].apply(lambda x: x.month)
z['day'] = z['date'].apply(lambda x: x.day)
z = z.loc[(z.month == 1) & (z.day == 6)]
z['count'] = z['transcript'].apply(lambda x: len(x.split()))
z
```

```python
max_count = res_df['partition_id'].value_counts().max()
```

```python
shows = res_df['partition_id'].unique()
```

```python
phrases = {
    'BBC News at One': 'good afternoon and welcome to the bbc news at one',
    'BBC News at Six': 'good evening and welcome to the bbc news at six',
    'BBC News at Ten': 'tonight at ten'
}
```

```python
counts = {'BBC News at Six': 0, 'BBC News at Ten': 0, 'BBC News at One': 0}
```

```python
rows = []
cols = ['date', 'source', 'transcript1', 'topic1', 'transcript2', 'topic2', 'transcript3', 'topic3', 'transcript4', 'topic4', 'transcript5', 'topic5']
for show in shows:
    df_show = res_df.loc[res_df.partition_id == show]
    i = 0
    
    phrase = phrases[df_show.iloc[0]['source']]
    
    while i < len(df_show):
        row = df_show.iloc[i]
        if phrase in row['transcript'].lower():
            break
        i += 1
    
    if i == len(df_show):
        i = 0
    
    while i < len(df_show):
        row = df_show.iloc[i]
        if len(row['transcript'].split()) > 50:
            break
        i += 1
        
    row = df_show.iloc[i]
    df_row = [row['date'], row['source']]
    
    for j in range(5):
        row = df_show.iloc[j+i]
        df_row.extend([len(row['transcript'].split()), row['topic']])
    rows.append(df_row)

res_df2 = pd.DataFrame(rows, columns = cols)
res_df2
```

```python
res_df2.to_csv('News_at_1_6_10_with_first_5_partitions_with_transcript_Jan_June_2016.csv')
```

```python
res_df2['count'] = 1
```

```python
topics = ['topic1', 'topic2', 'topic3', 'topic4', 'topic5']
programs = ['BBC News at One', 'BBC News at Ten', 'BBC News at Six']
total_counts = res_df2['source'].value_counts()
cols = ['source'] + topics

rows = []
for program in programs:
    row = [program]
    for topic in topics:
        counts = res_df2[['source', topic, 'count']].groupby(['source', topic], as_index=False).sum()
        counts = counts.loc[(counts.source == program) & (counts[topic] == 'Parliament, government and politics')]
        assert(len(counts) == 1)
        count = counts.iloc[0]['count']
        row.append(count/ total_counts[program])
    rows.append(row)
    
parliament_share_df = pd.DataFrame(rows, columns=cols)
parliament_share_df
```

```python
parliament_share_df.to_csv('parliament_position_share.csv', index=False)
```

```python
rows = []
cols = ['date', 'source']

for i in range(max_count):
    cols.append('word_count ' + str(i))

for show in shows:
    df_show = res_df.loc[res_df.partition_id == show]
    
    row = df_show.iloc[0]
    df_row = [row['date'], row['source']]
    df_row.extend([0] * max_count)
    i = 2
    for index, row in df_show.iterrows():
        df_row[i] = len(row['transcript'].split())
        i += 1
    rows.append(df_row)

res_df3 = pd.DataFrame(rows, columns = cols)
res_df3
```

```python
res_df3.to_csv('./partition_word_counts_News_at_One_Six_Ten.csv')
```

```python
rows = []
cols = ['date', 'source', 'topic', 'word_count']


for show in shows:
    df_show = res_df.loc[res_df.partition_id == show]
    df_show = df_show.reset_index(drop=True)
    
    max_count = 0
    max_count_index = 0
    
    flag = False
    i = 0
    for index, row in df_show.iterrows():
        _len = len(row['transcript'].split())
        
        if _len >= 1000:
            df_row = [row['date'], row['source'], row['topic'], _len]
            flag = True
            break
        else:
            if max_count < _len:
                max_count = _len
                max_count_index = index
        i += 1
    
    if flag:
        rows.append(df_row)
    else:
        row = df_show.iloc[max_count_index]
        df_row = [row['date'], row['source'], row['topic'], max_count]
        rows.append(df_row)

res_df4 = pd.DataFrame(rows, columns = cols)
res_df4
```

```python
res_df4.to_csv('News_at_One_Six_Ten_topic_with_first_partion_1000_words_July_2015_Dec_2015.csv')
```
