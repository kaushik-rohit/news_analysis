```python
import pandas as pd
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
```

```python
def get_bigrams(text):            
    stopW = stopwords.words('english')
    
    # To stem
    ps = PorterStemmer()

    # Tokenize
    
    bigrams = {}
    
    clean_transcript = text.lower()
    clean_transcript = word_tokenize(clean_transcript)
    # Remove digits
    clean_transcript = [i for i in clean_transcript if not re.match(r'\d+', i)]
    # Remove Stopwords and single characters
    clean_transcript = [i for i in clean_transcript if i not in stopW and len(i) > 1]
    # Stemming
    clean_transcript = [ps.stem(word) for word in clean_transcript]
    
    return ' '.join(clean_transcript)


def get_bigrams_count(text):            
    stopW = stopwords.words('english')
    
    # To stem
    ps = PorterStemmer()

    # Tokenize
    
    bigrams = {}
    
    clean_transcript = text.lower()
    clean_transcript = word_tokenize(clean_transcript)
    # Remove digits
    clean_transcript = [i for i in clean_transcript if not re.match(r'\d+', i)]
    # Remove Stopwords and single characters
    clean_transcript = [i for i in clean_transcript if i not in stopW and len(i) > 1]
    # Stemming
    clean_transcript = [ps.stem(word) for word in clean_transcript]
    
    no_words = len(clean_transcript)
    for word in range(no_words - 1):
        temp = clean_transcript[word] + '.' + clean_transcript[word + 1]
        if temp in bigrams:
            bigrams[temp] += 1
        else:
            bigrams[temp] = 1
    
    return bigrams

def get_short_sentences(transcript):
    sentences = nltk.sent_tokenize(transcript)
    n_sent = len(sentences)
    short_sentences = []
    lengths = []
    for i in range(n_sent):
        _len = len(sentences[i].split())
        lengths.append(_len)
        if _len >= 9:
            pass
        else:
            short_sentences.append(sentences[i])
            
    return ' '.join(short_sentences)

def contains_european(x):
    y = x.lower()
    words = y.split()
    flag = 'e.u' in words or 'eu' in words or 'europeanunion' in words
    words = [word for word in words if word not in stopW]
    bigrams = list(nltk.bigrams(words))
    bigrams = [bigram[0] + '.' + bigram[1] for bigram in bigrams]
    
    flag = flag or 'european.union' in bigrams or 'europe.union' in bigrams
    
    return 'EU' if flag else 'non_EU'
```

```python
bigrams_in_original_transcripts = {}
for index, row in raw_df.iterrows():
    bigrams = get_bigrams_count(row['Transcript'])
    
    for bigram, count in bigrams.items():
        if bigram in bigrams_in_original_transcripts.keys():
            bigrams_in_original_transcripts[bigram] += count
        else:
            bigrams_in_original_transcripts[bigram] = count
```

```python
def cmp_two_dictionaries(dct1, dct2):
    diff1 = {}
    diff2 = {}
    for key, value in dct1.items():
        if key not in dct2.keys():
            diff1[key] = value
        elif dct2[key] != value:
            diff1[key] = value
    for key, value in dct2.items():
        if key not in dct1.keys():
            diff2[key] = value
        elif dct1[key] != value:
            diff2[key] = value
    return diff1, diff2
```
