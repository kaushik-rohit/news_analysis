```python
import pandas as pd
from datetime import datetime, timedelta
from dateutil import parser
import re
import os
```

```python
def get_video_link(row):
    unavailable_link = row['Unavailable link']
    prog_id = re.findall('/[a-zA-Z0-9]+\?', unavailable_link)[0][1:-1]
    brod_id = re.findall('bcast=[a-zA-Z0-9]+', unavailable_link)[0].replace('bcast=', '')
    hour, minute = row['Time'].split(':')
    program_name = '-'.join(row['Program Name'].split(' '))
    source = '-'.join(row['Source'].split(' '))
    format_ = "%H:%M"
    time = (datetime.strptime(row['Time'], format_) - timedelta(hours=1)).strftime('%H%M%S')
    date = parser.parse(row['Date']).date()
    #print(time)
    return 'https://stream.learningonscreen.ac.uk/trilt-broadcasts/hls/{}-{}-{}T{}Z-{}.mp4/index.m3u8'.format(prog_id,
                                                                                                            program_name,
                                                                                                            date,
                                                                                                            time,
                                                                                                            source)
```

```python
data_path = './data/bbc'
```

```python
bbc_ids = os.listdir(data_path)
bbc_ids = [bbc_id for bbc_id in bbc_ids if not bbc_id.startswith('.DS')]
for bbc_id in bbc_ids:
    bbc_path = os.path.join(data_path, bbc_id)
    years = os.listdir(bbc_path)
    years = [year for year in years if not year.endswith('.pkl')]
    for year in years:
        year_path = os.path.join(bbc_path, year)
        no_transcript_path = os.path.join(year_path, 'no_transcripts')
        try:
            no_transcript_files = os.listdir(no_transcript_path)
            no_transcript_files = [f for f in no_transcript_files if f.endswith('.csv')]
        except:
            print('all transcript present for {} {}'.format(bbc_id, year))
            continue
        for no_transcript_file in no_transcript_files:
            print(no_transcript_file)
            no_transcript_file_path = os.path.join(no_transcript_path, no_transcript_file)
            df = pd.read_csv(no_transcript_file_path, index_col=0)
            df['video_link'] = df.apply(lambda x: get_video_link(x), axis=1)
            df.to_csv(no_transcript_file_path)
```
